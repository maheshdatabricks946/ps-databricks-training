```
list = range(100)
rdd = sc.parallelize(list)
rddmp = rdd.map(lambda x : (x,x**2))
aggrdd = rddmp.filter(lambda x : x[0]%2==0) 
final = aggrdd.collect()
for i in final:
    print(i)
```

```
%scala
val xRdd= sc.parallelize(Array(('B',5),('B',4),('A',3),('A',2),('A',1)))
val yRdd= xRdd.groupByKey()
println("x,"+xRdd.getNumPartitions)
println("y,"+yRdd.getNumPartitions)
println(xRdd.collect().mkString(","))
println(yRdd.collect().mkString(","))
```

```
%scala
val x= sc.parallelize(Seq(("math", 55),("math", 56),("english", 57),("english", 58),("science", 59),("science", 54)))
val y= sc.parallelize(Seq(("math", 60),("math", 65),("science", 61),("science", 62),("history", 63),("history", 64)))
val s = x.sortByKey()
val z = s.join(y)
//print(s.collect().mkString(","))
print(z.collect().mkString(","))
```
